````python
# Answer 6
import sklearn
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score
import sys
import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv('train.csv')
df = df.replace([np.inf, -np.inf], np.nan)
df = df.fillna(0)
#print(df)
X = df.drop(['SalePrice'],axis=1)
Y = df['SalePrice']
# Split the data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=50)
y_train = np.log10(y_train)
y_test = np.log10(y_test)
# scaling dependent variables to have zero mean and unit variance
y_train = sklearn.preprocessing.scale(y_train)
y_test = sklearn.preprocessing.scale(y_test)

# part 1
alphas =  10**np.linspace(10,-6,100)*0.5
lasso_alpha = 0
best_lasso_mae = sys.maxsize * -1
ridge_alpha = 0
best_ridge_mae = sys.maxsize * -1

for alpha in alphas:
    # ridge
    ridge = linear_model.Ridge(alpha)
    # find MAE
    ridge_mae = cross_val_score(ridge, X_train, y_train, scoring="neg_mean_absolute_error", cv=5)
    avgRidgeMae = np.sum(ridge_mae) / 5
    if avgRidgeMae > best_ridge_mae:
        best_ridge_mae = avgRidgeMae
        ridge_alpha = alpha
    
    # lasso 
    lasso = linear_model.Lasso(alpha)
    # find MAE
    lasso_mae = cross_val_score(lasso, X_train, y_train, scoring="neg_mean_absolute_error", cv=5)
    avgLassoMae = np.sum(lasso_mae) / 5
    if avgLassoMae > best_lasso_mae:
        best_lasso_mae = avgLassoMae
        lasso_alpha = alpha
    
print("The best alpha for ridge is " + str(ridge_alpha) + " with a negative average MAE of " + str(best_ridge_mae))
print("The best alpha for lasso is " + str(lasso_alpha) + " with a negative average MAE of " + str(best_lasso_mae))
    
    
    
    
# part 2
for alpha in alphas:
    lm = linear_model.Ridge(alpha)
    lm.fit(X_train, y_train)
    plt.plot(lm.coef_)
    plt.xlabel("Parameter/Feature")
    plt.ylabel("Coefficient")
    plt.title("Ridge Plot")
plt.show()
for alpha in alphas:
    lm = linear_model.Lasso(alpha)
    lm.fit(X_train, y_train)
    plt.plot(lm.coef_)
    plt.xlabel("Parameter/Feature")
    plt.ylabel("Coefficient")
    plt.title("Lasso Plot")
plt.show()    
print("As the alphas increase the relevant feature coefficients decrease. Since Lasso uses the absolute value and we are computing least squares, its descent is non-linear while Ridge is linear.")

# part 3
lm = linear_model.LinearRegression()
lm.fit(X_train, y_train)
lm_predictions = lm.predict(X_test)
lm_MAE = mean_absolute_error(y_test, lm_predictions)

ridgeModel = linear_model.Ridge(ridge_alpha)
ridgeModel.fit(X_train, y_train)
ridge_predictions = ridgeModel.predict(X_test)
ridge_MAE = mean_absolute_error(y_test, ridge_predictions)

lassoModel = linear_model.Lasso(lasso_alpha)
lassoModel.fit(X_train, y_train)
lasso_predictions = lassoModel.predict(X_test)
lasso_MAE = mean_absolute_error(y_test, lasso_predictions)

print("\nMAE for linear regression model: ", lm_MAE)
print("MAE for ridge model: ", ridge_MAE)
print("MAE for lasso model: ", lasso_MAE)

# part 4
lassoModel_2 = LassoCV(alphas=None, cv=10, max_iter=10000)
lassoModel_2.fit(X_train, y_train)
print("\nAlpha for lasso with 10-fold cross validation: ", lassoModel_2.alpha_)

lassoModel_3 = linear_model.Lasso(lassoModel_2.alpha_)
lassoModel_3.fit(X_train, y_train)
print("\nCoefficients for the best alpha: ", lassoModel_3.coef_)

# part 5
print("\nIf we don't scale before regularization, regularizing may have different effects on different features. Large feature values will be affected less compared to smaller feature values, however after regularizing, the effect of our hyperparameter will be even across all features.")
````
